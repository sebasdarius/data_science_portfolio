```python
import pandas as pd
import numpy as np
```

```python
all_nba = pd.read_csv("nba_data/all_nba_by_year.csv")
all_star_old = pd.read_csv("nba_data/basketball_player_allstar.csv")
mvp = pd.read_csv("nba_data/mvp_by_year.csv")
dpoy = pd.read_csv("nba_data/dpoy_by_year.csv")
advanced = pd.read_csv("nba_data/player_season_advanced.csv")
finals_mvp = pd.read_csv("nba_data/finals_mvp_by_year.csv")
champions = pd.read_csv("nba_data/nba_champions.csv")
all_star = pd.read_csv("nba_data/all_star_by_year.csv")
```

```python
all_nba
```

```python
advanced = advanced.groupby(["slug", "name", "year"], as_index=False).agg(np.sum)
```

```python
advanced = advanced[["slug", "name", 
          "value_over_replacement_player", "year"]]
```

```python
master_df = advanced
```

### MVP

```python
mvp = mvp[["Season", "Player"]]
```

```python
def fix_season(season):
    return int(season[:4]) + 1
```

```python
mvp["Season"] = mvp["Season"].apply(fix_season)
```

```python
mvp["Name"] = mvp["Player"].apply(lambda x: x.split("\\")[0])
mvp["Id"] = mvp["Player"].apply(lambda x: x.split("\\")[1])
mvp.drop("Player", axis=1, inplace=True)
```

```python
mvp["Count_mvp"] = 1
mvp.set_index(["Id", "Season"], inplace=True)
```

```python
mvp.head()
```

```python
master_df = master_df.join(mvp[["Count_mvp"]], on=["slug", "year"], how="left")
master_df["Count_mvp"] = master_df["Count_mvp"].fillna(0)
```

### All NBA

```python
def all_nba_names(name):
    if name[-2:] in [" C", " F", " G"]:
        return name[:-2]
    else:
        return name
```

```python
all_nba.rename(columns={"0":"Player"}, inplace=True)
all_nba.Player = all_nba.Player.apply(all_nba_names)
```

```python
all_nba["First_Team_All_NBA"] = (all_nba["Team"] == "1st").astype("int8")
all_nba["Second_Team_All_NBA"] = (all_nba["Team"] == "2nd").astype("int8")
```

```python
all_nba.set_index(["Player", "Season"], inplace=True)
```

```python
def fix_names(x):
    if x[-1] == "*":
        return x[:-1]
    else:
        return x
```

```python
master_df.name = master_df.name.apply(fix_names)
```

```python
master_df = master_df.join(all_nba[["First_Team_All_NBA", "Second_Team_All_NBA"]], on=["name", "year"], how="left")
```

```python
master_df.fillna(0, inplace=True)
```

### All Star

### Championships

```python
champions.head()
```

```python
champions["team_year"] = (champions["Year"].astype("str") + " " +
                              champions["Champion"])
```

```python
playoffs = pd.read_csv("nba_data/player_playoff_totals.csv")
```

```python
playoffs.head()
```

```python
def get_team(team):
    try:
        if str(team) == "nan":
            return "None"
        else:
            team = team[5:]
            team = team.split("_")
            team[0] = team[0][0] + team[0][1:].lower()
            team_name = team[0]
            for word in team[1:]:
                word = word[0] + word[1:].lower()
                team_name = team_name + " " + word
            return team_name
    except:
        print(team)
```

```python
playoffs.team = playoffs.team.apply(get_team)
```

```python
playoffs["team_year"] = (playoffs["year"].astype('int32').astype(
    'str')) + " " + playoffs["team"]
```

```python
playoffs["Championships"] = playoffs["team_year"].apply(lambda x: x in list(champions["team_year"])).astype("int8")
```

```python
playoffs.year = playoffs.year.astype('int32')
playoffs = playoffs[["slug", "year", "Championships"]]
```

```python
playoffs.set_index(["slug", "year"], inplace=True)
```

```python
master_df = master_df.join(playoffs, on=["slug", "year"], how="left").fillna(0)
```

```python
ids = pd.read_csv("bball_ref_hof_pred.csv", encoding='latin')['id']
```

```python
master_df["qualified"] = master_df.slug.apply(lambda x: x in list(ids))
```

```python
master_df = master_df[master_df["qualified"]==True]
```

```python
master_df
```

```python
player_ids = list(master_df.slug.unique())
```

```python
df = pd.DataFrame()
for player_id in player_ids:
    player_df = master_df[master_df["slug"] == player_id].sort_values("year")
    
    slug = player_df["slug"].iloc[0]
    name = player_df["name"].iloc[0]
    
    for i in range(len(player_df)):
        temp_series = player_df[:i+1]
        year = temp_series["year"].iloc[-1]
        temp_series = temp_series[["value_over_replacement_player", "Count_mvp", "First_Team_All_NBA",
                    "Second_Team_All_NBA", "Count_dpoy", "Count_finals_mvp", "All_stars",
                    "Championships"]].sum()
        temp_series["slug"] = slug
        temp_series["name"] = name
        temp_series["year"] = year
        
        df = df.append(temp_series, ignore_index=True)
```
