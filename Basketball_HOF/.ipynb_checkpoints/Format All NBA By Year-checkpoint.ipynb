```python
import pandas as pd
import numpy as np
```

```python
all_nba = pd.read_csv("../nba_data/all_nba_by_year.csv")
all_nba.head()
```

```python
pd.DataFrame(np.array(next(all_nba.iterrows())[1][3:]).reshape(-1,1))
```

```python
def fix_season(season):
    return int(season[:4]) +1
```

```python
df = pd.DataFrame()
for row in all_nba.iterrows():
    temp_data = row[1]
    if temp_data["Lg"] != "NBA":
        continue
    temp_df = pd.DataFrame(np.array(row[1][3:]).reshape(-1,1))
    temp_df["Season"] = fix_season(temp_data["Season"])
    temp_df["Team"] = temp_data["Tm"]
    df = df.append(temp_df, ignore_index=True)
df
```

```python
df.to_csv("../nba_data/all_nba_by_year.csv")
```
